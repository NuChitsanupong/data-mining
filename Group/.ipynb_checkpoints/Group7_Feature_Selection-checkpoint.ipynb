```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
dataset = pd.read_csv("s1_heart_data.csv")
dataset.head()
```

```python
df = dataset[['heart_rate_stop', 'heart_rate_start', 'rest_time', 'temp_at_start', 'temp_at_stop', 'avg_temp', 'speed_start', 'speed_past_60_seconds', 'speed_past_600_seconds', 'speed_past_3600_seconds', 'speed_stop', 'avg_speed', 'heart_rate_past_60_seconds', 'heart_rate_past_600_seconds', 'heart_rate_past_3600_seconds']]
df.head()
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df = df.dropna()
df
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
input_vars = list(df.columns[1:])
input_vars
```

```python
df.describe()
```

```python
df.hist(figsize=(16,9))
```

## min-max normalize

```python
df_dataset = df.iloc[:,1:].copy()
df_dataset.head()
```

```python
normalized_df_dataset=(df_dataset-df_dataset.min())/(df_dataset.max()-df_dataset.min())
normalized_df_dataset.head()
```

```python
df_dataset = pd.concat([df.iloc[:,:1],normalized_df_dataset], axis=1)
df_dataset.head()
```

## Forward Feature Selection

```python
df_forward = df_dataset.copy()
```

```python
# f_regression receives input features and output
#              and returns f-statistic and p-value for f-test 
# f-test is to used to test whether model fits data well or not 
# the more f-statistic (the lower p-value), the better

from sklearn.feature_selection import f_regression
fstat, pval = f_regression(df_forward[input_vars], df_forward.heart_rate_stop)
fstat
```

```python
pval
```

```python
f_value_threshold = 100
updated_vars = []
for i in range(0,len(input_vars)-1):
    if fstat[i] >= f_value_threshold:
        updated_vars.append(df_forward[input_vars].columns[i])
updated_vars
```

```python
X = df_forward[updated_vars]
y = df['heart_rate_stop']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=99)
lm = LinearRegression()
lm.fit(X_train, y_train)
y_pred = lm.predict(X_test)
[np.sqrt(metrics.mean_squared_error(y_test,y_pred)),metrics.r2_score(y_test,y_pred)]
```

## Random Forest

```python
df_random_forest = df_dataset.copy()
```

```python
# RandomForestRegressor receives parameters for random forest such as depth of tree, number of trees

# n_features = number of selected features

from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(max_depth=10,n_estimators=100)
model.fit(df_random_forest[input_vars],df_random_forest.heart_rate_stop)

n_features = 8

# Plot feature importance
features = df_random_forest[input_vars].columns
importances = model.feature_importances_
indices = np.argsort(importances)[-n_features:]  # sort top n features
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()

# Choose features with large feature importance
```

```python
important_features = [features[i] for i in indices]
important_features
updated_vars = []
for i in range(-1,-n_features-1,-1):
    updated_vars.append(important_features[i])
updated_vars
```

```python
X = df[updated_vars]
y = df['heart_rate_stop']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=99)
lm = LinearRegression()
lm.fit(X_train, y_train)
y_pred = lm.predict(X_test)
[np.sqrt(metrics.mean_squared_error(y_test,y_pred)),metrics.r2_score(y_test,y_pred)]
```
