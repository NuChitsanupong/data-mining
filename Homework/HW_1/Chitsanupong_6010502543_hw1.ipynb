## b. Show the distribution of HDSize. What type of its distribution? Normal or Not?

```python
plt.figure(figsize=(16,9))

dist_plot = sns.distplot(sales_data['HD Size (GB)'], hist=True, kde=True, rug=False)
dist_plot.set_xlabel("HD Size (GB)", fontsize=16)
dist_plot.tick_params(labelsize=12)
```

## c. Show the distribution of RetailPrice. What type of its distribution? Normal or Not? 

```python
plt.figure(figsize=(16,9))

dist_plot = sns.distplot(sales_data['Retail Price'], hist=True, kde=True, rug=False, bins=10)
dist_plot.set_xlabel("Retail Price", fontsize=16)
dist_plot.tick_params(labelsize=12)
```

## d. Find whether there ate outliers in RetailPrice. Show your method to determine outliers

```python
plt.figure(figsize=(16,9))

box_plot = sns.boxplot(x='Retail Price', data=sales_data)
box_plot.set_xlabel("Retail Price", fontsize=16)
box_plot.tick_params(labelsize=12)
```

 boxplot  Outlier  Upper Bound  3   Outlier  Lower Bound  5 

 Outlier  IQR (Interquartile Range)  Q3  Q1 (Q3 = third quartile, Q1 = first quartile) 

```python
Q1 = sales_data['Retail Price'].quantile(0.25)
Q3 = sales_data['Retail Price'].quantile(0.75)

# interquartile range (IQR)
IQR = Q3 - Q1
upper = Q3 + 1.5*IQR
lower = Q1 - 1.5*IQR
```

```python
print('Upper: '+str(upper))
print('Lower: '+str(lower))
```

 630  350  Outlier

## e. What are mean, median, mode of HDSize and RetailPrice

```python
mean_HD_size = sales_data['HD Size (GB)'].mean()
mean_retail_price = sales_data['Retail Price'].mean()

median_HD_size = sales_data['HD Size (GB)'].median()
median_retail_price = sales_data['Retail Price'].median()

mode_HD_size = sales_data['HD Size (GB)'].mode()[0]
mode_retail_price = sales_data['Retail Price'].mode()[0]
```

```python
print('mean HD Size: '+ str(mean_HD_size))
print('mean Retail Price: '+ str(mean_retail_price))

print('median HD Size: '+ str(median_HD_size))
print('median Retail Price: '+ str(median_retail_price))

print('mode HD Size: '+ str(mode_HD_size))
print('mode Retail Price: '+ str(mode_retail_price))
```

## f. What are SD, variance, and IQR of HDSize and RetailPrice. Explain their meaning.

```python
std_HD_size = sales_data['HD Size (GB)'].std()
std_retail_price = sales_data['Retail Price'].std()

var_HD_size = sales_data['HD Size (GB)'].var()
var_retail_price = sales_data['Retail Price'].var()

Q1_HD_size = sales_data['HD Size (GB)'].quantile(0.25)
Q3_HD_size = sales_data['HD Size (GB)'].quantile(0.75)
IQR_HD_size = Q3_HD_size - Q1_HD_size

Q1_retail_price = sales_data['Retail Price'].quantile(0.25)
Q3_retail_price = sales_data['Retail Price'].quantile(0.75)
IQR_retail_price = Q3_retail_price - Q1_retail_price
```

```python
print('SD HD Size: '+ str(std_HD_size))
print('SD Retail Price: '+ str(std_retail_price))

print('variance HD Size: '+ str(var_HD_size))
print('variance Retail Price: '+ str(var_retail_price))

print('IQR HD Size: '+ str(IQR_HD_size))
print('IQR Retail Price: '+ str(IQR_retail_price))
```

Variance    2

IQR  Interquartile range   : IQR = Q3 - Q1

## g. Create a bar chart, showing the average RetailPrice by store. Which store has the
## highest average? Which has the lowest?

```python
sales_data_mean_by_store = sales_data.loc[:,['Store Postcode','Retail Price']].groupby('Store Postcode').mean()
sales_data_mean_by_store
```

```python
plt.figure(figsize=(16,9))

bar_plot = sns.barplot(x=sales_data_mean_by_store.index, y=sales_data_mean_by_store['Retail Price'], data=sales_data_mean_by_store)
bar_plot.set_xlabel("Store Postcode", fontsize=16)
bar_plot.set_ylabel("Average Retail Price", fontsize=16)
bar_plot.tick_params(labelsize=12)
bar_plot.set_xticklabels(sales_data_mean_by_store.index, rotation=45)
bar_plot.set(ylim=(480, 500))
```

```python
sales_data_mean_by_store_sorted = sales_data_mean_by_store.sort_values(by='Retail Price', ascending=False)
sales_data_mean_by_store_sorted
```

```python
plt.figure(figsize=(16,9))

bar_plot = sns.barplot(x=sales_data_mean_by_store_sorted.index, y=sales_data_mean_by_store_sorted['Retail Price'], data=sales_data_mean_by_store_sorted)
bar_plot.set_xlabel("Store Postcode", fontsize=16)
bar_plot.set_ylabel("Average Retail Price", fontsize=16)
bar_plot.tick_params(labelsize=12)
bar_plot.set_xticklabels(sales_data_mean_by_store_sorted.index, rotation=45)
bar_plot.set(ylim=(480, 500))
```
