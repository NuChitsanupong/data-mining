```python
student_profile = pd.read_excel('university_translated.xls', sheet_name='STUDENT_PROFILE')
student_profile
```

```python
student_profile.DEPARTMENT_ID.replace(
    [1, 2, 3, 4, 5, 6], 
    ['Computer Engineering', 'Civil Engineering', 'Electrical Engineering', 'Chemical Engineering', 'Machanic Engineering', 'Environmental Engineering'], 
    inplace=True)

student_profile.rename({'DEPARTMENT_ID': 'DEPARTMENT'}, axis=1, inplace=True)

student_profile.head()
```

```python
student_profile.ZIP_CODE.replace(
    [1, 2, 3, 4, 5], 
    ['Bangkok', 'Central', 'Northen', 'Southern', 'Northeastern'], 
    inplace=True)

student_profile.rename({'ZIP_CODE': 'REGION'}, axis=1, inplace=True)

student_profile.head()
```

```python
student_grade_profile = pd.merge(student_profile, student_grade, on='STUDENT_ID')
student_grade_profile
```

```python
student_grade_profile['AGE_1_YEAR'] = pd.cut(
    x = student_grade_profile['AGE_1_YEAR'], bins=[15,18,21,23],                               
    labels = ["Low","Medium","High"])
student_grade_profile
```

```python
student_grade_profile['GRADE_204111'] = pd.cut(
    x = student_grade_profile['GRADE_204111'], bins=[-1,1.5,2.5,4],                               
    labels = ["Low","Medium","High"])

student_grade_profile['GRADE_204222'] = pd.cut(
    x = student_grade_profile['GRADE_204222'], bins=[-1,1.5,2.5,4],                               
    labels = ["Low","Medium","High"])

student_grade_profile['GRADE_204333'] = pd.cut(
    x = student_grade_profile['GRADE_204333'], bins=[-1,1.5,2.5,4],                               
    labels = ["Low","Medium","High"])

student_grade_profile['GRADE_204444'] = pd.cut(
    x = student_grade_profile['GRADE_204444'], bins=[-1,1.5,2.5,4],                               
    labels = ["Low","Medium","High"])

student_grade_profile['GRADE_204555'] = pd.cut(
    x = student_grade_profile['GRADE_204555'], bins=[-1,1.5,2.5,4],                               
    labels = ["Low","Medium","High"])

student_grade_profile['GRADE_204666'] = pd.cut(
    x = student_grade_profile['GRADE_204666'], bins=[-1,1.5,2.5,4],                               
    labels = ["Low","Medium","High"])

student_grade_profile
```

```python
student_grade_profile_G_3 = student_grade_profile[(student_profile.STATUS=='G') & (student_profile.GPA>=3.00)]
```

```python
student_grade_profile_G_3 = student_grade_profile_G_3[['STUDENT_ID', 'GENDER', 'AGE_1_YEAR', 'REGION', 'DEPARTMENT', 'GRADE_204111', 'GRADE_204222', 'GRADE_204333', 'GRADE_204444', 'GRADE_204555', 'GRADE_204666']]
student_grade_profile_G_3.reset_index(drop=True, inplace=True)
student_grade_profile_G_3
```
